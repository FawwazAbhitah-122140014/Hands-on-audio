```python
import os
import cv2
```

```python
import os

img_path = os.path.join(os.getcwd(), 'data', 'Ramona.png')
img2_path = os.path.join(os.getcwd(), 'data', 'Ramona.png')

# cheack file exists
assert os.path.exists(img_path), f"File {img_path} tidak ditemukan"
assert os.path.exists(img2_path), f"File {img2_path} tidak ditemukan"

print("Path gambar 1:", img_path)
print("Path gambar 2:", img2_path)
```

```python
#load gambar
img = cv2.imread(img_path)

print(f"img.shape: {img.shape}")  # (tinggi, lebar, jumlah channel)
plt.figure(figsize=(12, 6))
plt.imshow(img)
plt.axis('off')
plt.title('original image (open)')
```

```python
#Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12, 6))
plt.imshow(img_rgb)
plt.axis('off')
plt.title('original image (RGB)')
```

```python
#cara bikin kotak tengah dengan numpy
img_rgb_copy = img_rgb.copy()
height, width, _ = img_rgb_copy.shape
box_size = 100
start_y = (height - box_size) // 2
start_x = (width - box_size) // 2
img_rgb_copy[start_y:start_y + box_size, start_x:start_x + box_size] = [255, 0, 0]  # Merah
plt.figure(figsize=(12, 6))
plt.imshow(img_rgb_copy)
plt.axis('off')
plt.title('image with red box (numpy)')
```

```python
# border hitam dan merah
img_rgb_border = img_rgb.copy()
plt.figure(figsize=(12, 6))
plt.imshow(img_rgb_border)
plt.axis('off')
plt.title('image with border (numpy)')
```

```python
# Bikin histogram pada gambar di atas
colors = ('r', 'g', 'b')
plt.figure(figsize=(12, 6))
for i, color in enumerate(colors):
    hist = cv2.calcHist([img_rgb_border], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)
    plt.xlim([0, 256])
plt.title('Histogram for color scale picture (numpy)')
plt.xlabel('Pixel value')
plt.ylabel('Number of pixels')
plt.show()
```

```python
# ADD 15 PTS TO RED, Blue, Green PIXEL VALUES
img_rgb_bright = img_rgb.copy()
img_rgb_bright[:, :, 0] = np.clip(img_rgb_bright[:, :, 0] + 100, 0, 255)
img_rgb_bright[:, :, 1] = np.clip(img_rgb_bright[:, :, 1] + 50, 0, 255)
img_rgb_bright[:, :, 2] = np.clip(img_rgb_bright[:, :, 2] + 100, 0, 255)
plt.figure(figsize=(6, 6))
plt.imshow(img_rgb_bright)
plt.axis('off')
plt.title('brightened image (numpy)') 
```

```python
#Histogram of brightened image
plt.figure(figsize=(12, 6))
for i, color in enumerate(colors):
    hist = cv2.calcHist([img_rgb_bright], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)
    plt.xlim([0, 256])
plt.title('Histogram for brightened picture (numpy)')
plt.xlabel('Pixel value')
plt.ylabel('Number of pixels')
plt.show()
```
