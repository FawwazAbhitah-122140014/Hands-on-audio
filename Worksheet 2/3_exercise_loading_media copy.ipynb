# Pendahuluan

## Apa itu Data Multimedia?

Data multimedia adalah informasi yang dikodekan dalam berbagai format untuk merepresentasikan dunia nyata:

- **Audio (1D)**: Sinyal satu dimensi yang berubah terhadap waktu
  - Contoh: musik, suara, speech
  - Representasi: amplitudo vs waktu
  
- **Gambar (2D)**: Matriks nilai intensitas dalam ruang dua dimensi
  - Contoh: foto, ilustrasi, grafik
  - Representasi: intensitas pixel pada koordinat (x,y)
  
- **Video (2D + Waktu)**: Rangkaian frame (gambar) yang ditampilkan berurutan
  - Contoh: film, rekaman, animasi
  - Representasi: frame berubah terhadap waktu dengan frame rate tertentu

**Instruksi:**
Baca video dengan OpenCV dan tampilkan metadata:
- Resolusi (Width √ó Height)
- Frame rate (fps)
- Jumlah total frame
- Durasi (detik)
- Klasifikasi resolusi (HD, Full HD, 4K, dll.)

**Analisis yang diperlukan:**
Jelaskan pentingnya parameter-parameter tersebut untuk analisis video atau aplikasi tertentu.

```python
# Membaca video dan menampilkan metadata dasar
file_path = os.path.join("data", "bocchi 3D.mp4") 
cap = cv2.VideoCapture(file_path)

if not cap.isOpened():
    print("Gagal membuka video!")
else:
    # Ambil metadata
    width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps    = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / fps if fps > 0 else 0

    # Jenis-Jenis resolusi
    if width >= 3840 and height >= 2160:
        res_class = "4K (Ultra HD)"
    elif width >= 1920 and height >= 1080:
        res_class = "Full HD (1080p)"
    elif width >= 1280 and height >= 720:
        res_class = "HD (720p)"
    elif width >= 640 and height >= 480:
        res_class = "SD (480p)"
    else:
        res_class = "Low Resolution"

    # Tampilkan hasil
    print("=== Informasi Metadata Video ===")
    print(f"Resolusi       : {width} x {height}")
    print(f"Frame rate     : {fps:.2f} fps")
    print(f"Jumlah frame   : {total_frames}")
    print(f"Durasi         : {duration:.2f} detik")
    print(f"Jenis Resolusi : {res_class}")

cap.release()
```

**Analisis parameter video**

Saat melakukan analasis video resolusi berfungsi menentukan kualitas dari video tersebut, lalu frame rate berfungsi untuk mendektesi gerakan video, jumlah frame adalah total frame atau banyaknya gambar pada video tersebut, lalu durasi adalah lama waktu dari video tersebut. dan klasifikasi resolusi berfungsi untuk mengetahui kualitas dari video tersebut.

### C3. TODO: Tampilkan 3 Frame (Awal‚ÄìTengah‚ÄìAkhir)

**Instruksi:**
Ambil dan tampilkan 3 frame representatif:
- Frame pertama (index 0)
- Frame tengah (index ~total_frame/2)
- Frame terakhir (index total_frame-1)
- **Konversi BGR‚ÜíRGB** sebelum ditampilkan
- Subplot dengan judul frame dan timestamp

**Analisis yang diperlukan:**
Deskripsikan perbedaan visual antar frame dan apa yang dapat dipelajari dari sampel frame ini.

```python
# Membaca video dan menampilkan 3 frame representatif
file_path = os.path.join("data", "bocchi 3D.mp4") 
cap = cv2.VideoCapture(file_path)

if not cap.isOpened():
    print("Gagal membuka video!")
else:
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    # Menentukan index frame (awal, tengah, akhir)
    frame_indices = [0, total_frames // 2, total_frames - 1]
    frames = []

    for idx in frame_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)  # pindah ke frame tertentu
        ret, frame = cap.read()
        if ret:
            # Konversi BGR ‚Üí RGB untuk matplotlib
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            timestamp = idx / fps
            frames.append((frame_rgb, idx, timestamp))

    cap.release()

    # menampilkan isi frame dan informasi
    plt.figure(figsize=(15, 5))
    for i, (frame_rgb, idx, timestamp) in enumerate(frames):
        plt.subplot(1, 3, i+1)
        plt.imshow(frame_rgb)
        plt.axis("off")
        plt.title(f"Frame {idx}\nWaktu: {timestamp:.2f} detik")

    plt.suptitle("3 Frame Representatif (Awal ‚Äì Tengah ‚Äì Akhir)", fontsize=14)
    plt.show()
```

**Analisis perbedaan visual antar frame**

-*Frame pertama*

Pada frame pertama menampilkan character 1 yang berekpresikan ceria
